```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import os
os.chdir("../")
```

```python
df = pd.read_csv('data/rsna-bone-age/training/train_df.csv')
```

```python
sample_image = df.path[10]
```

```python
def visualize(path):
    img = plt.imread(path)
    img = np.clip(img, 0, 0.8, out=img)
    plt.imshow(img, cmap='gray')
    plt.colorbar()
    plt.show()
```

```python
visualize(sample_image)
```

```python
# contrast limited adaptive histogram equalization
from skimage import exposure
```

```python
def CLAHE(path):
    img = plt.imread(path)
    img = exposure.equalize_adapthist(img)
    plt.imshow(img, cmap='gray')
    plt.title("CLAHE augmentation")
    plt.colorbar()
    plt.show()
```

```python
CLAHE(sample_image), visualize(sample_image)
```
