```python
import glob
import os
import cv2
import matplotlib.pyplot as plt
```

```python
os.chdir("../")
```

```python
paths = glob.glob("data/Mexico_private_dataset/additional/*")
for path in paths:
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    plt.imshow(img)
    plt.show()
    y_crop = input()
    box = (0, 0, y_crop, img.shape[1])
    cropped = img[box[0]:box[2], box[1]:box[3]]
    cv2.imwrite(os.path.join("data/Mexico_private_dataset/cropped/",os.path.basename(path)), cropped)
```

```python
img_num = 85
img = cv2.imread(f"data/Mexico_private_dataset/cropped/{img_num}.png", cv2.IMREAD_UNCHANGED)
flipped = cv2.flip(img, 1)
cv2.imwrite(os.path.join(f"data/Mexico_private_dataset/cropped/{img_num}.png"), flipped)
```
